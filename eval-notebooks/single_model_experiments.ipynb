```python
from evaluator import Evaluator
import pandas as pd
```

```python
csv = pd.read_csv("exper_data.csv")
```

```python
models = [
    "OpenAI: GPT-5.4",
    "Anthropic: Claude Sonnet 4.6",
    "Google: Gemini 3 Flash Preview",
    "DeepSeek: DeepSeek V3.2"
]
```

```python
batch_number = 5
batch_size = 50

start = batch_size * (batch_number - 1)
end = min(batch_size * batch_number, len(csv))

rows = []

i = 1
for _, row in csv.iloc[start:end].iterrows():
    evaluator = Evaluator(*models)
    result = [{"model": model, "response": "ERROR"} for model in models]

    try:
        result = await evaluator.query_models(models, [row["question"]] * len(models))
        print(f"({i}/{batch_size}) Done with question: {row['id']}")
        i += 1
    except Exception as e:
        print(str(e))

    for dic in result:
        rows.append({
            "id": row["id"],
            "question": row["question"],
            "model": dic["model"],
            "response": str(dic["response"])
        })

df = pd.DataFrame(rows)
```

```python
df.to_parquet(
    f"single_model_experiment_results/model_responses_batch_{batch_number}.parquet",
    index=False
)
df.to_csv(
    f"single_model_experiment_results/model_responses_batch_{batch_number}.csv",
    index=False
)
```
