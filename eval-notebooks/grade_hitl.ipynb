# Human-in-the-Loop Adaptive Rubrics (GradeHITL)

This notebook implements a **human-in-the-loop adaptive grading pipeline** (`GradeHITL`).

**Goal**: Start with an initial rubric, let the LLM grade automatically, and only ask a human for help on ambiguous or low-confidence cases. Use those interventions to refine the rubric so the model’s grading converges toward human-level behavior.

**High-level loop (per batch of responses)**:
- **1) Grade**: LLM grades each response using the current rubric and outputs scores + justification + a confidence estimate.
- **2) Flag**: The system identifies low-confidence or borderline cases according to thresholds (e.g., confidence < 0.6 or scores near a pass/fail cutoff).
- **3) Ask human**: For flagged cases, the LLM generates targeted clarification questions for the educator and (optionally) proposes rubric tweaks.
- **4) Update rubric**: The educator answers, and the system folds these answers back into the rubric prompt/policy.
- **5) Regrade & iterate**: Regrade affected items with the updated rubric, and repeat over time as new edge cases appear.

Take a look at `grade_hitl.py` for the implementation of this evaluation method.
